### Ziel
Darstellung der Powerfunktion als Wahrscheinlichkeit $P_p(X \in K)$ in Abhängigkeit vom wahren Parameter $p$, wobei $K$ der Ablehnungsbereich von $H_0$ ist.

### Eingabe
- n: Stichprobenumfang  
- p_0    : Modell unter H₀  
- alpha : Signifikanzniveau  
- p_wahr : Beispielparameter (optional)

### Ausführen
Beim ersten Start bitte **Run All Cells** ausführen.

Danach: Die Zellen *Modell* sowie *Plot und ggf. Grafik speichern* ausführen: **Run (▶)**.

```python
# ======== Imports ============
from tests.model.binomial import BinomialModel
from tests.geometry.construct_rejection_region import two_sided_equal_tails
from tests.plots.plot_power import plot_power_curve
from tests.plots.plot_model import ModelPlotStyle
```

```python
# =============== Modell =========================
# Modellparameter
n = 100
p0 = 0.5

# Setzung
alpha = 0.2
p_wahr = 0.6
```

```python
# ========= Plot und ggf. Grafik speichern =========

model_null = BinomialModel(n=n, p=p0)
R = two_sided_equal_tails(model_null, alpha=alpha)

style = ModelPlotStyle(
    power_ref_width=2, power_ref_style="--", 
    power_point_size=20.0,
)

plot_power_curve(
    model_factory=lambda p: BinomialModel(n=n, p=p),
    rejection_region=R,
    p0=p0, n=n, alpha=alpha,
    p_star=p_wahr, p_step=0.005,
    style=style,
    #save=f"grafik_power_n{n}_p{p0:.2f}_ptrue{p_wahr:.2f}.pdf"   
);
```

# Start – Aktivitäten (Überblick)

### Überblick

Dieses Repository ist so aufgebaut, dass Sie mit möglichst wenig Code arbeiten:

- **notebooks/** enthält die Aktivitäten (hier arbeiten Sie).
- **lib/** enthält die fertigen Referenzprogramme (nicht verändern).

Wählen Sie eine Aktivität und ändern Sie nur die Parameter.

## Aktivitäten
- Aktivität 01: Konfidenzellipse (Wilson-KI)
- Aktivität 02: Prognoseintervalle (relativ/absolut)
- Aktivität 03: Simulation (Überdeckungsrate)
- Aktivität 04: Ablehnungsbereich (z-Test)

## Start
Öffnen Sie eine Aktivität und führen Sie die Zellen von oben nach unten aus.

### Simulation: Überdeckungsrate von Konfidenzintervallen

Ein einzelnes Konfidenzintervall liefert **keine** Aussage darüber, ob es den wahren Parameter enthält.
Das Sicherheitsniveau $\gamma$ ist eine **Verfahrenseigenschaft** und wird durch Wiederholung sichtbar.

In dieser Aktivität werden viele Stichproben simuliert und die Überdeckungsrate des Wilson-Verfahrens geschätzt.

```python
import setup_path
from lib.intervalle import (
    plot_ci_simulation,
    CISimConfig,
    CISimStyle,
)
```

### 1. Parameter setzen
- `n`: Stichprobengröße
- `p_true`: wahrer Parameterwert
- `m`: Anzahl Simulationen (bei Binder eher moderat)

```python
cfg = CISimConfig(
    # ---- Modell -----
    n=50,
    p_true=0.40,   # p_true ist eine Setzung innerhalb eines Modells
   
    # ---- Setzung ----
    gamma=0.95,
    m=50,
    seed=42,
    x_min=0.0,
    x_max=0.8,
)

style = CISimStyle(
    color_mode="coverage",   # alle Intervalle gleich, sonst ""coverage"
    show_stats=True,       # SuS sollen selbst zählen (False)
    show_grid=False,
    
)

plot_ci_simulation(cfg, style, save=None);
```

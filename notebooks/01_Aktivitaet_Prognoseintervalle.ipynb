## Aktivität 02 – Prognoseintervalle (relativ vs. absolut)

### Prognoseintervalle (relativ und absolut)

Hier wird ein Prognoseintervall für eine Binomialverteilung betrachtet.

- **Relativ:** für $H=X/n$ (Anteil)  
- **Absolut:** für $X$ (Anzahl)

Beide Darstellungen gehören zusammen, haben aber unterschiedliche Achsen und Interpretationen.

```python
import setup_path
from lib.intervalle import plot_pi, PIModel, IntervalGeometry, CIStyle
```

### 1. Parameter setzen

```python
model = PIModel(
    p=0.4, 
    n=50, 
    gamma=0.95
    )

geometry = IntervalGeometry(
    x_min=0.0,
    x_max=1.0
)

style = CIStyle(
    curve_upper="gray",
    curve_lower="gray",
    area_color="lightgray",
    area_alpha=0.25,
    interval_color="tab:blue",
    helper_lines="gray",
    grid=False,
)

plot_pi(model, geometry, style,
       absolute=True,
       save=None
       );
```

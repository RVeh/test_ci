### Sicherheitswahrscheinlichkeit - was ist das? 

Der Begriff **Sicherheitswahrscheinlichkeit** wird in diesem Zusammenhang oft benutzt. Geeigneter ist der Begriff **Verfahrenswahrscheinlichkeit**.

Die Aussage: Es wird für alle $p$ garantiert, dass 
  $$
    P_p\bigl( KI(X) \ni p \bigr) \ge \gamma
   $$
gilt, wird mit Simulationen illustriert, indem viele Realisationen desselben Modells erzeugt und ausgewertet werden.

#### Wie geschieht das?

Ausgehend von einem festen Modellparameter $p$ (eigentlich unbekant) werden viele Stichproben
$X_1, X_2, \dots, X_m$ simuliert.
Zu jeder Realisation wird ein Konfidenzintervall berechnet.
Gezählt wird, wie oft der feste Parameter $p$ im zufälligen Intervall $KI(X)$ liegt.

#### Was haben wir in der Realität?

- Wir haben nur eine Stichprobe.
- Wir fragen:
    - Was hätte auch passieren können?
    - Was müsste ein Verfahren leisten, dem wir vertrauen wollen?
   

#### Wo steckt denn der Zufall? 

- Der Zufall steckt nicht im Parameter.
- Er steckt nicht im Vertrauen.
- Er steckt nicht im konkreten Intervall.
- Er steckt **im Gedanken der Wiederholung.**
